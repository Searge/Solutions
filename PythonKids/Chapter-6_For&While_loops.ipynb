# Цикли та їх застосування
## Застосування <span style="color:red">«FOR»</span> циклів

Замість повторення будь-якої операції, наприклад, 5 разів _можна_ скористатися циклом `for`:

````python
for x in range(0, 5):
    print('Привіт!')
````

**Функцію** `range` (діапазон) можна застосовувати для створення переліку чисел від першого до числа перед останнім числом (бо знову пайтонівський список з індексом).
Вона не стоврює сам список чисел, а повертає _ітератор_ — такий _тип об'єктів_, який придуманий спеціально для роботи з циклами.
_Грубо кажучи_ в попередньому циклі `for` каже, що для `x` ми повертаємо діапазон чисел від $1$ (`0`) до $5$ (`4`). Тобто:

- рахуємо з `0` та зупиняємося, не досягнувши `5`.
- для кожного порахованого числа зберігати значення у змінній `x`
- далі Python запускає код блоку (`print('Привіт!')`).

Якщо об'єднаємо ф-цію `range` з ф-цією `list`, ми отримаємо список чисел:

````python
print(list(range(10, 20)))
````

````python
print(list(range(0, 5)))
````

**Порахуємо наші привітання:**

````python
for x in range(0, 5):
    print('Привіт, %s!' % x)
````

Змінна `x` це така ж змінна, як і без циклу `for`:

````python
x = 0
print('Привіт, %s!' % x)
````

Щоб написати `for`-цикл не обов'язково користуватися тільки ф-ціями `range` та `list`.
Можна також скористатися уже створеним нами списком, напр. списком покупок із [Розділу 3](https://nbviewer.jupyter.org/github/Searge/Solutions/blob/master/PythonKids/Chapter-3_lists_tuples_dicts.ipynb):

````python
список_чарівника = [
    'павукові лапки', 
    'жаб\'ячий палець',
    'око тритона',
    'крило кажана',
    'масло зі слизняка',
    'ведмежа відрижка'
]

for i in список_чарівника:
    print(i)
````

Це ще один спосіб сказати: «Для кожного елементу зі `список_чарівника` збережи значення у змінній `i`, а тоді: надрукуй вміст цієї змінної».
Знову ж таки, якщо ми позбудемося `for`-циклу, то будемо змушенні друкувати кожен пункт окремо:

````python
список_чарівника = [
    'павукові лапки', 
    'жаб\'ячий палець',
    'око тритона',
    'крило кажана',
    'масло зі слизняка',
    'ведмежа відрижка'
]

print(список_чарівника[0])
print(список_чарівника[1])
print(список_чарівника[2])
print(список_чарівника[3])
print(список_чарівника[4])
print(список_чарівника[5])
````

````python
# Ось складніший цикл з двома блоками коду:

великіволохатіштани = ['великі', 'волохаті', 'штани']
for i in великіволохатіштани:
    print(i)
    for j in великіволохатіштани:
        print('—', j)
````

Після створення списку `великіволохатіштани` код проходить по елементам списку та друкує кожен з них, однак, після друку кожного значення зі змінної `i`, він проходить ще один цикл присвоюючи значення змінній `j` та друкує її.

В Розділі 2 ми рахували скільки монет буде в кінці року:

````python
20 + 10 * 365 - 3 * 52
````

> У ньому позначено 20 знайдених монет, плюс 10 чарівних, помножених на 365 днів, мінус 3 монети на тиждень які вкрала ворона 🐦

За допомогою цикла ми можемо побачити, як наша скрабниця збільшується щотижня. Для цього змінемо значення змінної `чарівні_монети`, щоб вона позначала загальну кількість монет на тиждень, тобто ${10}$ _монет_ × ${7}$ _днів у тижні_ і створемо ще одну зміну `монети`:

````python
знайдені_монети = 20
чарівні_монети = 70
вкрадені_монети = 3
монети = знайдені_монети

for тиждень in range(1, 53):
    монети = монети + чарівні_монети - вкрадені_монети
    print('Тиждень %s = %s' % (тиждень, монети))
````

Рядок `монети = монети + чарівні_монети - вкрадені_монети` означає: «Заміни вміст змінної `монети` на поточну кількість моїх монет, плюс кількість монет, які я створив цього тижня».
По суті, знак дорівнює (=) — це такий фрагмент коду (оператор), який любить командувати і каже:
«Спочатку порахуй, що там треба, праворуч, а тоді збережи результат на потім, під тією назвою, що ліворуч».

## До слова про цикли…
Також є цикл `while` (поки). `For` — цикл **певної довжини**, тоді як `while` – це цикл, яким користуються, коли наперед не знають, скільки обчислень має бути (**якої довжини він має бути**).

Уяви 20 сходинок, на які ти легко можеш піднятися застосувавши цикл `for`:

````python
for сходинка in range(0, 20):
    print(сходинка)
````

А тепер уяви сходи, що ведуть на гору:

````python
сходинка = 0
while сходинка < 10000:
    print(сходинка)
    if втомився == True:
        break
    elif погана_погода == True:
        break
    else:
        сходинка = сходинка + 1
````

> Цей код для демонстрації принципу, спроба його запустити викличе помилку, т.я. змінні `втомився` та `погана_погода` не означенні.

Етапи роботи циклу `while`:

1. Перевірити умову
2. Виконати код у блоці
3. Повторити

Часто `while` створюють з кількома умовами, а не лише однією:


````python
x = 45
y = 80

while x < 50 and y < 100:
    x = x + 1
    y = y + 1
    print(x, y)
````

В цьому прикладі ми створили змінну `x` зі значенням `45` і змінну `y` зі значенням `80`.
Цикл _перевіряє_ чи `x`**менше** _за_ `50`, **та** чи `y` **менше** _за_ `100`.
Доки обидві умови задовільняються (_справедливі_) наступні рядки виконуються, додаючи по `1` до кожної змінної.
А тоді вже друкується результат.

Ще одним поширеним застосуванням `while` є створення **напівскінченних циклів**.
Цикл цього типу може повторюватися **нескінченно**, але таки припиняє працювати, коли в коді стається щось, що його перериває:

````python
while True:
    тут багато коду
    тут багато коду
    тут багато коду
    if якась_умова == True:
        break
````

Тут умова `while` — просто `True`, а вона **завжди** правдива, тож код виконуватиметься **завжди** (_вічно_), аж поки `якась_умова` не виявиться `True`.

### №1. Цикл «Привіт»
Спробуй здогадатися/зрозуміти, що робить цей код:

````python
for x in range(0, 20):
    print('привіт %s' % x) # друкує привіт № з/п
    if x < 9: # перериває цикл не дійшовши 9-ти 
        break
    # Ага, навіть не 9-ти, а й 1-ки, бо умова, що x менше 9-ти правильна з самого початку!
````

Якби ми зараз були на Місяці, твоя вага складала б `16.5%` від твоєї ваги на Землі.

````python
weight = 65
x = 1

for overweight in range(0, 15):
    weight = weight + 1
    on_moon = weight * 0.165
    print('Рік %s — %s кг.' % (x, on_moon))
    x = x + 1
````
