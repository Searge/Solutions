## Копіювання за допомогою модуля `copy`

Зазвичай, пишучи програму, ти створюєш нові об'єкти, але іноді корисно створити копію об’єкта, а тоді скористатися цією копією, щоб створити новий об’єкт, а надто, коли процес творення об’єкта складається з кількох кроків.

Скажімо, ми маємо клас `Тварина`, з ф-цією `__init__`, що приймає параметри: `вид, кількість_ніг` та `забарвлення`.

```python
class Тварина:
    def __init__(self, вид, кількість_ніг, забарвлення):
        self.вид = вид
        self.кількість_ніг = кількість_ніг
        self.забарвлення = забарвлення
```

Ми можемо створити новий об’єкт у класі `Тварина`, застосувавши наступний код:

```python
# Створимо гіпогрифа на ім’я гаррі

гаррі = Тварина('гіпогриф', 6, 'рожевий')
```

А якщо нам потрібна зграя гіпогрифів на шістьох ногах?

```python
import copy
гаррієт = copy.copy(гаррі)

print(гаррі.вид, гаррієт.вид)
```

```python
керрі = Тварина('химера', 4, 'в зелені крапочки')
біллі = Тварина('велетень', 0, 'з візерунком')
мої_тварини = [гаррі, керрі, біллі]
більше_тварин = copy.copy(мої_тварини)

print(більше_тварин[0].вид, більше_тварин[1].вид)
```

```python
мої_тварини[0].вид = 'упир'

print(більше_тварин[0].вид)
```

Ф-ція `copy` створює _поверхневу_ копію — вона не копіює об’єкти всередині скопійованих об’єктів.

<mark>Таким чином</mark>,  якщо ми додамо нову тварину до першого списку, вона не з’явиться у другому.

```python
саллі = Тварина('сфінкс', 4, 'піщаний')
мої_тварини.append(саллі)

print(len(мої_тварини), len(більше_тварин))
```

А от інша ф-ція і модулі `copy` — `deepcopy`, створює копії всіх об’єктів усередині тих об’єктів, які копіюються. Коли застосовуємо ф-цію `deepcopy` для копіювання списку `мої_тварини`, ми отримуємо новий список із копіями усіх його об’єктів. 
У результаті зміни, внесені до одного з наших початкових об’єктів класу `Тварина`,  не вплинуть на об’єкти в новому списку.

```python
більше_тварин = copy.deepcopy(мої_тварини)
```

```python
мої_тварини[0].вид = 'змій'

print(мої_тварини[0].вид, більше_тварин[0].вид)
```
