# Повторне використання за доп. ф-цій та модулів

## Користування функціями

У [попередньому розділі](https://nbviewer.jupyter.org/github/Searge/Solutions/blob/master/PythonKids/Chapter-6_For%26While_loops.ipynb) ми скористалися ф-ціями `range` та `list`, щоб змусити Python лічити

```python
list(range(0, 5))
```

**_Функції_** — це фрагмент коду, які кажуть Python виконати якусь дію, це _ще один спосіб_ повторного використання коду.

У написанні простих програм ф-ції можуть стати у нагоді, а в скалдніших, як от ігри — просто _необхідними_.

## Складові ф-ції
_Функція_ складається з трьох частин: <mark style="background-color: LightBlue">назви</mark>, <mark style="background-color: LemonChiffon">параметрів</mark> та <mark>тіла</mark>:

```python
def скажи_привіт(як_мене_звати):
    print('Привіт, %s' % як_мене_звати)
```

Назва цієї ф-ції — `скажи_привіт`.
Вона має єдиний параметр — `як_мене_звати`, а її тіло — це блок коду відразу за рядком, який починається з `def` (скорочення від **define** — _визначити, означити_).
_Параметр_ — це змінна, яка існує тільки тоді, коли ф-цією користуються.

Ф-цію можна викликати, вказавши її назву і поставивши дужки навколо значення параметра:

```python
скажи_привіт('Мері')
```

Вона може приймати скільки завгодно параметрів, а не лише один:

```python
def скажи_привіт(ймення, прізвище):
    print('Привіт, %s %s' % (ймення, прізвище))
    
#  викликаючи ф-цію, розділяємо значення комою:

скажи_привіт('Мері', 'Сміт')
```

Також спочатку можна створити кілька змінних, а тоді вже, скориставшись ними, викликати ф-цію:

```python
моє_ймення = 'Джо'
моє_прізвище = 'Робертсон'

скажи_привіт(моє_ймення, моє_прізвище)
```

Часто використовують для повернення значення, застосовуючи інструкцію `return` (_повернути_).
Наприклад, ти можеш написати ф-цію, що рахуватиме, скільки грошей тобі вдалося відкласти:

```python
def заощадження(кишенькові, заробітки, витрати):
    return кишенькові + заробітки - витрати
```

Тут вона приймає три параметри, потім: додає перші два та віднімає останній.
Результат обчислення можна присвоїти змінній, або можна вивести на екран:

```python
print(заощадження(10, 10, 5))
```

## Змінні та область видимості
Змінну, яка перебуває всередині тіла ф-ції не можна використовувти **ще раз**, коли ф-ція вже виконала свою дію, бо вона існує _лише_ у **межах ф-ції**.
Це називають _область видимості_

Розгляньмо просто ф-цію без параметрів, яка користується кількома змінними:

```python
def тест_змінних():
    перша_змінна = 10
    друга_змінна = 20
    return перша_змінна * друга_змінна

print(тест_змінних)
print(тест_змінних())
```

Якщо ми спробуємо надрукувати зміст будь-якої змінної всередині ф-ції, то отримаємо повідомлення про помилку:

```python
print(перша_змінна)
```

Натомість, можна визначити змінну до створення ф-ції:

```python
ще_одна_змінна = 100
def тест_змінних_2():
    перша_змінна = 10
    друга_змінна = 20
    return перша_змінна * друга_змінна * ще_одна_змінна

print(тест_змінних_2())
```

Уявімо, що ти будуєш космічний корабель із пресованих бляшанок.
Нехай то буде 2 бляшанки на тиждень.
Щоб закінчити фюзеляж, тобі знадобиться 500 бляшанок.

Створимо ф-цію, щоб показати скільки бляшанок ми спресували щотижня протягом року:

```python
def будівництво_корабля(бляшанки):
    всього_бляшанок = 0
    for тиждень in range(1, 53):
        всього_бляшанок = всього_бляшанок + бляшанки
        print('Тиждень %s = %s бляшанок' % (тиждень, \
                                           всього_бляшанок))

будівництво_корабля(2)
```

## Застосування модулів
_Модулі_ застосовуються для групування ф-цій, змінних _та інших_ елементів мови програмування у більші, потужніші програми.

Наприклад, якщо ти пишеш гру і хочеш, щоб світ змінювався у реальному часі, ти можеш застосуватим вбудованний модуль `time`:

```python
import time

print(time.asctime())
```

А тепер спробуємо ввести свій вік в програму за допомогою модуля `sys` (**system**) який містить засоби для спілкування з самою системою Python.

В середині модуля є спеціальний об'єкт `stdin` (**standart input** — _стандартний ввід_), що надає ф-цію `readline` (_прочитати рядок_).

В [5-му розділі](https://nbviewer.jupyter.org/github/Searge/Solutions/blob/master/PythonKids/Chapter-5_If_that_then_that.ipynb) ми застосовували умовну інструкцію для перевірки віку. Тепер її можна зробити інтерактивною:

```python
import sys


def дурний_жарт_про_вік():
    print('Скільки тобі років?')
    вік = int(sys.stdin.readline())
    if вік >= 10 and вік <= 13:
        print('Скільки буде 13 + 49 + 84 + 155 + 97')
        print('Багато!')
    else:
        print('Що?')
```

```python
дурний_жарт_про_вік()
```

<mark>Помилка</mark> означає, що _«ти не можеш конвертувати порожню стрічку (**string**) `''` в число (**int**)»_. Тому я використаю `input()`, який веде себе більш передбачуванно і не треба нічого імпортувати.

```python
def дурний_жарт_про_вік():
    print('Скільки тобі років?')
    вік = int(input())
    if вік >= 10 and вік <= 13:
        print('Скільки буде 13 + 49 + 84 + 155 + 97')
        print('Багато!')
    else:
        print('Що?')
```

```python
дурний_жарт_про_вік()
```

### №1. Вага на Місяці
У [Розділі №6](https://nbviewer.jupyter.org/github/Searge/Solutions/blob/master/PythonKids/Chapter-6_For%26While_loops.ipynb) одним із завдань було створити `for`-цикл, щоб визначити твою вагу на Місяці, його легко перетворити на ф-цію, а потім викликати за допомогою `вага_на_місяці(30, 0.25)`

```python
def вага_на_місяці(вага, кг):
    for рік in range(1, 16):
        вага = вага + кг
        вага_на_місяці = вага * 0.165
        print('Рік %s твоя вага %s' % (рік, вага_на_місяці))
```

```python
вага_на_місяці(30, 0.25)
```

```python
вага_на_місяці(65, 1)
```

### №2. Вага на Місяці за конкретні роки
Зміни створену ф-цію так, щоб можна було вирахувати вагу за різні періоди, наприклад за 5 років.
Пильнуй за тим, щоб вона прийсала **три аргументи**: _початкову вагу; вагу, додану щороку та кількість років_ — `вага_на_місяці(90, 0.25, 5)`

```python
def вага_на_місяці(вага, кг, роки):
    for рік in range(1, роки + 1):
        вага = вага + кг
        вага_на_місяці = вага * 0.165
        print('Рік %s твоя вага %s' % (рік, вага_на_місяці))
```

```python
вага_на_місяці(90, 0.25, 5)
```

### Програма вираховування ваги на Місяці
Замість простої ф-ції, у яку значення передаються як параметри, ти можеш створити міні-програму, яка просить вваести значення через `sys.stdin.readline()`.
У цьому випадку ти можеш викликати ф-цію взагалі без будь-яких параметрів:

`вага_на_місяці()`

Ф-ція відобразить повідомлення, в якому попросить ввести початкову вагу, а потім інше…

<samp>
    Введи свою поточну вагу на Землі
    45
    Введи, наскільки збільшуватиметься твоя вага щороку
    0.4
    Введи кількість років
    12
</samp>

```python
import sys


def вага_на_місяці(вага, кг, роки):
    print('Введи свою поточну вагу на Землі')
    вага = float(sys.stdin.readline())
    print('Введи, наскільки збільшуватиметься твоя вага щороку')
    кг = float(sys.stdin.readline())
    print('Введи кількість років')
    роки = int(sys.stdin.readline())
    for рік in range(1, роки + 1):
        вага = вага + кг
        вага_на_місяці = вага * 0.165
        print('Рік %s твоя вага %s' % (рік, вага_на_місяці))
        

вага_на_місяці()
```
