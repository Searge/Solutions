# Стрічки

```python
fred = '''Як тиранозаври сплачують рахунки?
Тиранозаврячами чеками'''
print(fred)
```

```python
мійрахунок = 1000
повідомлення = 'Мені вдалося здобути %s очок'
print(повідомлення % мійрахунок)
```

```python
числа = 'Що сказало число %s числу %s? \n— Гарний пояс!!'
print(числа % (0, 8))
```

### Множення стрічок

```python
пробіли = ' ' * 25
print('%s 🚀 Полетіли!!!' % пробіли)
```

# Списки

```python
список_чарівника = [
    'павукові лапки', 
    'жаб\'ячий палець',
    'око тритона',
    'крило кажана',
    'масло зі слизняка',
    'луска змії'
]

print(список_чарівника)
```

```python
print(список_чарівника[2])

'''3-й елемент списку 🖕!!!'''
```

>Для того, щоб змінити елемент списку, вказуємо його номер в індексі [рохуємо з нуля]

```python
список_чарівника[2] = 'язик равлика'
print(список_чарівника)
```

>За допомогою двокрапки і кв. дужок [:] можна показати частину списку

```python
print(список_чарівника[2:5])

'''Показуються елементи 2,3,4 (п'ять не включене в список)'''
```

```python
числа_та_рядки = ['Скільки', 'буде', 6, 'на', 8, '?']

# суміш стрічок та чисел
print(числа_та_рядки)
```

```python
# Вкладенні списки

числа = [1, 2, 3, 4]
стрічки = [
    'Я',
    'забив',
    'мій',
    'палець',
    'і',
    'він',
    'тепер',
    'болить'
]

мій_список = [числа, стрічки]
print(мій_список)
```

## Додавання до списку

```python
список_чарівника.append('ведмежа відрижка')
print(список_чарівника)
```

```python
# Додаємо іще!

список_чарівника.append('мандрагора')
список_чарівника.append('болиголов')
список_чарівника.append('болотний газ')

print(список_чарівника)
```

## Видалення зі списку

```python
del список_чарівника[5] # шостий елемент!
print(список_чарівника)
```

## Арифметика зі списками

```python
список1 = [1, 2, 3, 4]
список2 = ['я', 'перечипився', 'упав', 'і', 'забився']
print(список1 + список2)
```

```python
# ми також можемо об'єднати два списки та присвоїти результат в іншій змінній

список3 = список1 + список2
print(список3)
```

```python
# А ще можна помножити список на число
# я, правда, поки не розумію навіщо ¯\_(ツ)_/¯ 

print(список1 * 5)
```

## Кортежі

```python
фібоначчі = (0, 1, 1, 2, 3)
print(фібоначчі)
```

**Кортеж** — відрізняється від списку тим, що після створення його неможливо змінити. 
Наприклад, якщо ми спробуємо замінити перше значення в кортежі `фібоначчі` числом `4`, то дістанемо повідомлення про помилку.

> **Списки** використовують `[квадратні]` дужки, **кортежі** — `(звичайні)`.

```python
фібоначчі[0] = 4
```

## Словники

Кожен елемент у **словниках** має `ключ` та відповідне `значення` через двокрапку.

> Елементи у **словнику** беруться у `{фігурні}` дужки.
>> також, **словники** не можна об'єднувати, як **списки** за допомогою арифм. оператора **➕**

```python
улюблений_спорт = {
    'Ральф Вільямс' : 'футбол',
    'Майкл Тіппет' : 'баскетбол',
    'Едвард Ельгар' : 'бейсбол',
    'Ребекка Кларк' : 'нетбол',
    'Етель Сміт' : 'бадмінтон',
    'Френк Брідж' : 'реґбі',
}

print(улюблений_спорт['Ребекка Кларк'])
```

```python
# Видалити елемент можна тією ж командою, в [квадратних] дужках просто вказують ключ

del улюблений_спорт['Етель Сміт']
print(улюблений_спорт)
```

```python
# Так само, щоб замінити значення у словнику ми вказуємо ключ 
# і через присвоєння змінної = 'нове значення'

улюблений_спорт['Ральф Вільямс'] = 'хокей на льодові'
print(улюблений_спорт)
```

### № 1. Улюблене
Склади список своїх улюблених хобі і дай йому назву змінної `ігри`.
А тепер склади список своїх улюблених страв і назви цю змінну `їжа`
Об'єднай два списки і назви результат `улюблене`.
У кінці надрукуй змінну `улюблене`:

```python
ігри = ['Team Frotress 2', 'CS GO', 'Python 🐍']
їжа = ['🍰', '🍗', '🍔', '🍤']
улюблене = [ігри + їжа]

print(улюблене)
```

### №2. Підрахунок воїнів
На кожному даху 3-х будівель ховалося по 25 ніндзь, а в 2-х тунелях — по 40 самураїв, скільки ніндзь та самураїв збиралися вступити у битву?

```python
ніндзь = 25
самураїв = 40
print(ніндзь + самураїв)
```

### №3. Вітання!
Створи дві змінні, одну для збереження твого імені, а другу — для прізвища. А тепер створи стрічку з тимчасовими замінниками (`%s`) і застосуй підстановку, щоб вставити своє ім'я в повідомлення.

```python
name = str(input())
hi = "Привіт, %s!"
print(hi % name)
```
