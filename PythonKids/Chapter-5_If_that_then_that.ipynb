## Інструкції «<span style="color:red">if</span>» that then «<span style="color:red">else</span>»

```python
вік = 23
if вік > 20: # умовна конструкція складається з ключового слова if умови та двокрапки:
    print('Оце стариган!') # Якщо відповідь True, тоді усі команди в блоці виконаються
    print('Що ти тут робиш?')
    print('Чому б тобі не зробити щось корисне?')
```

```python
вік = 10
if вік >= 10:
    print('Ти занадто старий для моїх жартів')
```

```python
print('Хочеш почути брудний жарт?')
вік = int(input('Вкажи вік:'))
if вік == 12:
    print('Порося впало в багнюку')
else:
    print('Тсс. Це секрет')
```

## Інструкції «<span style="color:red">якщо</span>» та «<span style="color:red">або якщо</span>»

```python
# elif скорочено від else if

if вік == 10:
    print('Як називається занадто засмагла малина?')
    print('Ожина!')
elif вік == 11:
    print('Що зелена виноградина сказала синій?')
    print('Дихай! Дихай!')
elif вік == 12:
    print('Що 0 сказав до 8?')
    print('Привіт, хлопці!')
elif вік == 13:
    print('У якому стані може перебувати вода?')
    print('У твердому, рідкому і газоподібному.')
else:
    print('Що?')
```

## Об'єднання умов

Умови можна об'єднувати, застосовуючи ключові слова `and` та `or`, що робить код коротшим та простішим.

```python
if вік == 10 or вік == 11 or вік == 12 or вік == 13:
    print('Скільки буде 13 + 49 + 84 + 155 + 97')
    print('Багато!')
else:
    print('Що?')
```

**Скоротимо** іще раз цей приклад, застосувавши ключове слово `and` разом із оператором _більше або дорівнює_ `>=`, чи _менше або дорівнює_ `<=`.

```python
if вік >= 10 and вік <= 13:
    print('Скільки буде 13 + 49 + 84 + 155 + 97')
    print('Багато!')
else:
    print('Що?')
```

## Змінні без значення
У **Python** порожнє значення називається `None` (ніщо), воно відрізняється від значення нуля, бо це відсутність будь-якого значення, а не число із значенням `0`.

```python
порожня_змінна = None
print(порожня_змінна)
```

Присвоєння _ніякого_ значення змінним зручно з того боку, що при створенні нової програми так легше бачити назви усіх змінних, що будуть пізніше застосовуватися.

Інструкції `if` теж можна налаштувати на перевірку _відсутності_ значення:

```python
if порожня_змінна == None:
    print('Змінна «порожня_змінна» — не має значення для мене')
```

_Це може стати у пригоді,_ коли ти хочеш вирахувати значення змінної, якщо воно ще не було вирахуване.

## Різниця між стрічками та числами

Комп'ютер сприймає `10` та `"10"`, як різні дані. Перше то числа `int()`, друге то текстова стрічка `str()`:

```python
вік = 10
if вік == 10:
    print('Як найкраще розмовляти з монстром?')
    print('Здалеку!')
```

```python
# Але якщо присвоїти стрічку:
вік = '10'
if вік == 10:
    print('Як найкраще розмовляти з монстром?')
    print('Здалеку!')
```

В цьому випадку команда `print` не запуститься, бо **Python** не вважає `'10'` за число, але ми можемо використати _магічну_ функцію `int` (**integer** — ціле число):

```python
конвертований_вік = int(вік)
```

```python
if конвертований_вік == 10:
    print('Як найкраще розмовляти з монстром?')
    print('Здалеку!')
```

Але якщо спробувати конвертувати число з десятковою **крапкою** (з рухомою комою), то отримаємо помилку, бо **функція** `int` очікує ціле число:

```python
вік = '10.5'
конвертований_вік = int(вік)
```

Літерал `literal` у програмуванні — це запис значення, який комп'ютер сприймає «буквально». Напр.: `10` — літерал, що явно означає $10$; назви змінних, з іншого боку, не є літералами, бо вони лише вказують на присвоєні їм значення.

Щоб виправити `ValueError` (помилку значення) треба замість `int` скористатися **функцією** `float` (яка працює з десятковими дробами):

```python
вік = '10.5'
конвертований_вік = float(вік)
```

```python
print(конвертований_вік)
```
