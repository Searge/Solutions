### Функція <mark>abs</mark>
Повертає _абсолютне значення_ (<a href="https://uk.wikipedia.org/wiki/Модуль_(математика)">величину</a>) числа, тобто — значення числа без його знаку.
Напр., абсолютне значення числа 10 є $10$; абсолютне значення числа -10 є $10$:

```python
print(abs(10))
```

```python
print(abs(-10))
```

Нею можна скористатися, щоб вирахувати абсолютну кількість рухів персонажа у грі, незалежно від того, в якому напрямку він рухається (+5 праворуч; -5 ліворуч).

Або ж у настільній грі, де треба кинути два кубика, а тоді пересунути свою фішку на к-сть кроків, яка дорівнює сумі чисел на кубиках в будь-якому напрямку.
Отже, якщо ми зберігатимемо к-сть кроків у змінній, ми зможемо визначити, чи персонаж рухається.

```python
кроки = -3

if abs(кроки) > 0:
    print('Персонаж рухається!')
```

```python
# без цієї ф-ції код виглядав би так:

if кроки < 0 or кроки > 0:
    print('Персонаж рухається!')
```

```python
if кроки != 0:
    print('Персонаж рухається!')
```

### Функція <mark>bool</mark>
Це скорочення від *[Boolean](https://uk.wikipedia.org/wiki/Булева_функція)*, вона бере один параметр і повертає `True` чи `False` в залежності від його значення.
Якщо застосовувати `bool` для чисел, то `0` повертається як `False`, а решта чисел — як `True`.

```python
print(bool(0))
```

```python
print(bool(1));print(bool(1123.23));print(bool(-500))
```

У стрічках `False`, тільки якщо стрічка пуста `''`, у всіх інших випадках — `True`.

```python
print(bool(None));print(bool(''))
```

```python
print(bool(' '));print(bool('a'));print(bool("Hello world!"))
```

А також для списків, кортежів та словників, якщо вони не мають значення:

```python
мій_дурний_список = []
print(bool(мій_дурний_список))

мій_дурний_список = ['д', 'у', 'р', 'н', 'и', 'й']
print(bool(мій_дурний_список))
```

```python
# Перевірка форми:

year = input('Рік народження: ')
if not bool(year.rstrip()):
    print('Ти маєш ввести рік свого народження')
```

```python
year = input('Рік народження: ')
if not year.rstrip():
    print('Ти маєш ввести рік свого народження')
```

Python автоматично переводить умовну інструкцію `if` в _Булеве значення_.
Ф-ція `rstrip()` — видаляє пробіли та натиснення <kbd>⏎</kbd> в кінці стрічки.
Сама інструкція каже: «Зроби це, якщо ф-ція **не** поверне значення `True`».

### Ф-ція <mark>dir</mark>
Скорочення від _directory_ (довідник), повертає інформацію про будь-яке значення.
Іншими словами, вона в алф. порядку перелічує нам ф-ції, які можна застосовувати з цим значенням.

```python
dir(['список'])
```

```python
dir(1)
```

```python
dir(None)
```

Ф-ція `dir` працює майже з усім, але іноді її результат можна ігнорувати, т.я. вона також повертає особливі ф-ції з якими працює сама мова Python (ті що поч. та зак. двома нижніми підкресленнями).

Якщо об'єкт, що передається ф-ції `dir` параметром, має атрибути, які не є ф-ціями, вони теж потраплять у цей список.

Також ця ф-ція може бути корисною, коли в тебе є змінна і ти хочеш швидко визначити, що з нею можна зробити.
Напр. запустити їїи,стуючись змінною `попкорн`, яка містить значення стрічку, ти отр. список ф-цій, наданих класом `str`:

```python
попкорн = ' Я люблю попкорн!'

dir(попкорн)
```

На цьому етапі ми скористаємося ф-цією `help`, щоб отримати короткий опис будь-якої ф-ції у списку:

```python
help(попкорн.upper)
```

* `(...)` означають, що це вбудованна ф-ція класу `str` і, _в цьому випадку_, не приймає ніяких параметрів.
* `->` означає, що вона повертає стрічку `str` (імплементує).
* _Повертає копію S, переведену у верхній регістр_.

### Ф-ція <mark>eval</mark>
Скорочення від _evaluate_ (_обчислити_) бере за параметр стрічку та виконує її так, наче то вираз Python. Вона працює тільки з простими виразами, а ті, що складаються з понад одного рядка (напр. інструкції `if`), зазвичай не обчислюються.

```python
eval('print("Wow!")')

eval('(5 + 4 + (2 - (3 - (6 + (4 / 5))))) / (3 * (6 - 2) * (2 - 7))')
```

```python
eval('if True: print("Це точно не вийде")')
```

Її зазвичай використовують для того, щоб, напр., написати простий калькулятор, який читатиме математичні вирази, а тоді обчислюватиме відповіді. Позаяк, для того, щоб Python прочитав числа та оператори введенні як стрічка (через `input()`), треба їх конвертувати в відповідні дані, а `eval` робить цю конвертацію сам за замовчуванням.

```python
calc = input('Введи обчислення: ')
eval(calc)
```

### Ф-ція <mark>exec</mark>
Вона схожа на `eval`, от тільки нею можна користуватися для запуску складніших програм. А ще `eval` повертає значення, а `exec` — ні.

```python
моя_маленька_програма = '''print('шинка')
print('бутерброд')'''

exec(моя_маленька_програма)
```

Цю ф-цію можна застосовувати для запуску міні-програм прямо з файлів, що може стати у нагоді під час написання довгих, складних додатків.

### Ф-ція <mark>float</mark>
Перетворює стрічку/число на _число з рухомою крапкою (комою)_, або ж на десятковий дріб (_дійсне число_ [$\mathbb{R}$](https://uk.wikipedia.org/wiki/Дійсне_число))

```python
число = float(input('Введи будь-яке число: '))
print(число)
```

### Ф-ція <mark>int</mark>
Перетворює стрічку/число на ціле число [$\mathbb{Z}$](https://uk.wikipedia.org/wiki/Цілі_числа), а це означає, що частина після крапки просто зникає (а не округлюється).

```python
print(int(число))
```

```python
# Перетворення стрічки
int('12')
```

> Це працює <mark>тільки якщо</mark> в стрічці знаходиться **ціле**, а не **дійсне** число. Перетворення стрічки з десятковим дробом викличе помилку:

```python
int('3.14159265359')
```

### Ф-ція <mark>len</mark>
Повертає довжину (_length_) об'єкта, або кількість символів у стрічці.

```python
len('довжина цієї стрічки')
```

У разі застосування до списку, кортежу чи словника, вона повертає кількість елементів у ньому:

```python
кортеж_істот = ('єдиноріг', 'циклоп', 'фея', 'ельф', 'дракон', 'троль')

словник_ворогів = {'Бетмен' : 'Джокер',
                  'Супермен' : 'Лекс Лютор',
                  'Спайдермен' : 'Зелений Гоблін',}

print(len(кортеж_істот))
print(len(словник_ворогів))
```

Дуже добре її використовувати в циклах. Напр., щоб вивести на екран індекси елементів списку.

```python
фрукт = ['яблуко', 'банан', 'апельсин', 'манго']
кількість = len(фрукт)
for x in range(0, кількість):
    print('Фрукт під №%s це %s' % (x, фрукт[x]))
```

### Ф-ції <mark>max</mark> та <mark>min</mark>
`max` повертає найбільший елемент стрічки, списку, кортежу і т.д.

<mark>Більшою літерою</mark>, *буде та, яка йде **пізніше** в* <mark>[лексографічному](https://uk.wikipedia.org/wiki/Лексикографічний_порядок) порядкові</mark>, при чому українські літери `є, і, ї, ґ` йдуть в кінці. Також рядкові літери більші від великих.

```python
укр = 'г, ґ, є, Г, Ґ, Є'
print(max(укр))
```

```python
абетка_рядкові = 'а, б, в, г, ґ, д, е, є, ж, з, и, і, ї, й, к, л, м, н, о, п, р, с, т, у, ф, х, ц, ч, ш, щ, ь, ю, я, '
абетка_великі = абетка_рядкові
абетка = абетка_великі.upper() + абетка_рядкові

print(абетка)
```

```python
print(sorted(абетка.split(', ')))
```

```python
print(max(абетка))
```

```python
print(min(абетка))
```

```python
список_літер = абетка.split(', ')

x = len(список_літер) - 1
del список_літер[x]

print(список_літер)
```

```python
print(max(список_літер));print(min(список_літер))
```

```python
print(sorted(список_літер))
```

```python
# Порівняння чисел

print(max(10, 300, 450, 50, 90))

числа = [5, 4, 10, 30, 22]
print(min(числа))
```

```python
# Гра на загадування найменшого числа.

вгадай_це_число = 61
здогади_гравців = [12, 15, 70, 45]
if max(здогади_гравців) > вгадай_це_число:
    print('Ха! Ви всі програли!')
else:
    print('Ви виграли!')
```

### Ф-ція <mark>range</mark>
Застосовується переважно у `for`-циклах.
Перші два параметри, надані ф-ції `range(x, y)`, називаються початком та кінцем.
Власне, ф-ція `range` повертає спец. об'єкт під назвою _<mark>ітератор</mark>_, який повторює якусь дію певну к-сть разів.

```python
for x in range(0, 5):
    print(x)
```

Ітератор можна перетворити на список (заст. ф-цію `list`).

```python
print(list(range(0, 5)))
```

До ф-ції `range(x, y, z)` можна додати і третій аргумент, крок. Якщо він не заданий, то за замовчуванням задається значення рівне $1$.

```python
рахуй_по_два = list(range(0, 30, 2))
print(рахуй_по_два)
```

```python
зворотній_відлік = list(range(40, 10, -2))
print(зворотній_відлік)
```

### Ф-ція <mark>sum</mark>
Додає елементи списку і повертає суму.

```python
мій_список_чисел = list(range(0, 500, 50))

print(мій_список_чисел, '=', sum(мій_список_чисел))
```
