# Як користуватися класами та об'єктами

## Розділення предметів на класи

Ми можемо застосовувати _класи_ для організації фрагментів коду, як от в `turtle` — рухи вперед, назад, праворуч, ліворуч — це все ф-ції класу `Pen`.

Тепер створимо такий самий набір класів, як на діаграмі:

````python
class Предмети:
    pass
````

<mark>`pass`</mark> застосовується тоді, коли ми хочемо означити клас, чи ф-цію, але на разі не хочемо заповнювати детал (що вона робить).

Це звичайний спосіб написання програм. Часто програмісти створюють класи з ф-ціями, які нічого не роблять, для того, щоб розібратися, що мають робити класи, перш ніж переходити до роботи над деталями окремих ф-цій.

````python
class Неживі(Предмети):
    pass
class Живі(Предмети):
    pass

class Тротуари(Неживі):
    pass

class Тварини(Живі):
    pass
class Ссавці(Тварини):
    pass
class Жирафи(Ссавці):
    pass
````

## Додавання об'єктів до класів
Нехай в нас є жирафа Реджинальд, щоб вказати, що він є об'єктом класу `Жирафи` (чи _екземпляром класу_), нам необхідно присвоїти його за доп. оператора `=`:

````python
реджинальд = Жирафи()
````

Цей код каже: «створити об'єкт в класі `Жирафи` та присвоїти його, як значення, змінній `реджинальд`».
<mark>Так само як і ф-ція</mark>, назва класу супроводжується дужками.

## Визначення функцій класів

Для ф-цій класу (або методів класу) ми маємо робити відступ, як і для будь-якого іншого блоку коду в Python. Також вони мають приймати <mark>принаймі один параметр</mark> в середині класу (зазвичай його називають `self`).

## Додавання класових характеристик, як ф-цій
Ми можемо додати _характеристики_ до кожного класу, щоб описати його детальніше і визначити, що він може робити.
<mark>Характеристика</mark> — це властивість, яку поділяють усі члени класу (та його нащадки). Їх можна вважати діями, або ф-ціями — тим, що можуть робити об'єкти того класу.

````python
class Тварини(Живі):
    def дихати(self):
        pass
    def рухатися(self):
        pass
    def харчуватися_їжею(self):
        pass

class Ссавці(Тварини):
    def годувати_дитинчат_молоком(self):
        pass

class Жирафи(Ссавці):
    def їсти_листя_з_дерев(self):
        pass
````

## Навіщо користуватися класами та об'єктами?
Через те, що `реджинальд` об'єкт, ми можемо викликати ф-ції, властиві його класові (`Жирафи`) та його батьківським класам.
Ми викликаємо ф-ції об'єкта, застосовуючи оператор крапку (`.`) та назву ф-ції.

````python
реджинальд.рухатися()
реджинальд.їсти_листя_з_дерев()
````

Змінемо ф-ції, щоб вони виводили на екран свою дію:

````python
class Тварини(Живі):
    def дихати(self):
        print('дихає')
    def рухатися(self):
        print('рухається')
    def харчуватися_їжею(self):
        print('харчується їжею')

class Ссавці(Тварини):
    def годувати_дитинчат_молоком(self):
        print('годує дитинчат')

class Жирафи(Ссавці):
    def їсти_листя_з_дерев(self):
        print('їсть листя')
````

````python
гарольд = Жирафи()

реджинальд.рухатися()
гарольд.їсти_листя_з_дерев()
````

## Інші корисні властивості об'єктів та класів
З класами та об'єктами <mark>легко групувати ф-ції</mark>, вони також дуже зручні, коли треба розділити програму на менші частини.

Уяви, що нам хотілось би розширити деякі класи, що ми їх створили у цьому розділі, але в нас багато роботи і ми хочемо, щоб нам допомогли друзі.
Ми могли б розділити роботу з написання коду так, що одна людина працює над класом `Тварини`, друга — `Ссавці`, ще одна — над класом `Жирафи`.

## Виклик ф-ціями інших ф-цій
Щоб якась ф-ція у класі `Жирафи` викликала ф-цію рухатися, ми натомість застосуємо параметр `self`.
Параметр `self` — це спосіб для однієї ф-ції в класі викликати іншу ф-цію.
Напр. ми додали ф-цію `шукати_їжу` до класу `Жирафи`:

````python
class things:
    pass
class alive(things):
    pass

class animals(alive):
    def move(self):
        print('moving')
    def eat(self):
        print('харчується їжею')
        
class mammals(animals):
    def breastfeeding(self):
        print('feeding')

class giraffe(mammals):
    def find_food(self):
        self.move()
        print('Я знайшов їжу!')
        self.eat()
    def eat_leaves(self):
        self.eat()
    def dancing(self):
        self.move()
        self.move()
        self.move()
        self.move()
````

````python
harrold = giraffe()
````

````python
harrold.find_food()
````
