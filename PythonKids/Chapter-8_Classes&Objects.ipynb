# Як користуватися класами та об'єктами

## Розділення предметів на класи

Ми можемо застосовувати _класи_ для організації фрагментів коду, як от в `turtle` — рухи вперед, назад, праворуч, ліворуч — це все ф-ції класу `Pen`.
<img alt="mermaid diagram" src="img/diagram-Living-n-Non-Living-Things.svg" width="200">
 
Тепер створимо такий самий набір класів, як на діаграмі:

```python
class Предмети:
    pass
```

<mark>`pass`</mark> застосовується тоді, коли ми хочемо означити клас, чи ф-цію, але на разі не хочемо заповнювати детал (що вона робить).

Це звичайний спосіб написання програм. Часто програмісти створюють класи з ф-ціями, які нічого не роблять, для того, щоб розібратися, що мають робити класи, перш ніж переходити до роботи над деталями окремих ф-цій.

```python
class Неживі(Предмети):
    pass
class Живі(Предмети):
    pass

class Тротуари(Неживі):
    pass

class Тварини(Живі):
    pass
class Ссавці(Тварини):
    pass
class Жирафи(Ссавці):
    pass
```

## Додавання об'єктів до класів
Нехай в нас є жирафа Реджинальд, щоб вказати, що він є об'єктом класу `Жирафи` (чи _екземпляром класу_), нам необхідно присвоїти його за доп. оператора `=`:

```python
реджинальд = Жирафи()
```

Цей код каже: «створити об'єкт в класі `Жирафи` та присвоїти його, як значення, змінній `реджинальд`».
<mark>Так само як і ф-ція</mark>, назва класу супроводжується дужками.
