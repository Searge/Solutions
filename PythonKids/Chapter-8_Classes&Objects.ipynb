Тепер створимо такий самий набір класів, як на діаграмі:

````python
class Предмети:
    pass
````

<mark>`pass`</mark> застосовується тоді, коли ми хочемо означити клас, чи ф-цію, але на разі не хочемо заповнювати детал (що вона робить).

Це звичайний спосіб написання програм. Часто програмісти створюють класи з ф-ціями, які нічого не роблять, для того, щоб розібратися, що мають робити класи, перш ніж переходити до роботи над деталями окремих ф-цій.

````python
class Неживі(Предмети):
    pass
class Живі(Предмети):
    pass

class Тротуари(Неживі):
    pass

class Тварини(Живі):
    pass
class Ссавці(Тварини):
    pass
class Жирафи(Ссавці):
    pass
````

<mark>Характеристика</mark> — це властивість, яку поділяють усі члени класу (та його нащадки). Їх можна вважати діями, або ф-ціями — тим, що можуть робити об'єкти того класу.

````python
class Тварини(Живі):
    def дихати(self):
        pass
    def рухатися(self):
        pass
    def харчуватися_їжею(self):
        pass

class Ссавці(Тварини):
    def годувати_дитинчат_молоком(self):
        pass

class Жирафи(Ссавці):
    def їсти_листя_з_дерев(self):
        pass
````

## Навіщо користуватися класами та об'єктами?
Через те, що `реджинальд` об'єкт, ми можемо викликати ф-ції, властиві його класові (`Жирафи`) та його батьківським класам.
Ми викликаємо ф-ції об'єкта, застосовуючи оператор крапку (`.`) та назву ф-ції.

````python
реджинальд.рухатися()
реджинальд.їсти_листя_з_дерев()
````

Змінемо ф-ції, щоб вони виводили на екран свою дію:

````python
class Тварини(Живі):
    def дихати(self):
        print('дихає')
    def рухатися(self):
        print('рухається')
    def харчуватися_їжею(self):
        print('харчується їжею')

class Ссавці(Тварини):
    def годувати_дитинчат_молоком(self):
        print('годує дитинчат')

class Жирафи(Ссавці):
    def їсти_листя_з_дерев(self):
        print('їсть листя')
````

````python
гарольд = Жирафи()

реджинальд.рухатися()
гарольд.їсти_листя_з_дерев()
````

## Інші корисні властивості об'єктів та класів
З класами та об'єктами <mark>легко групувати ф-ції</mark>, вони також дуже зручні, коли треба розділити програму на менші частини.

Уяви, що нам хотілось би розширити деякі класи, що ми їх створили у цьому розділі, але в нас багато роботи і ми хочемо, щоб нам допомогли друзі.
Ми могли б розділити роботу з написання коду так, що одна людина працює над класом `Тварини`, друга — `Ссавці`, ще одна — над класом `Жирафи`.

## Виклик ф-ціями інших ф-цій
Щоб якась ф-ція у класі `Жирафи` викликала ф-цію рухатися, ми натомість застосуємо параметр `self`.
Параметр `self` — це спосіб для однієї ф-ції в класі викликати іншу ф-цію.
Напр. ми додали ф-цію `шукати_їжу` до класу `Жирафи`:

````python
class things:
    pass
class alive(things):
    pass

class animals(alive):
    def move(self):
        print('moving')
    def eat(self):
        print('харчується їжею')
        
class mammals(animals):
    def breastfeeding(self):
        print('feeding')

class giraffe(mammals):
    def find_food(self):
        self.move()
        print('Я знайшов їжу!')
        self.eat()
    def eat_leaves(self):
        self.eat()
    def dancing(self):
        self.move()
        self.move()
        self.move()
        self.move()
````

````python
harrold = giraffe()
````

````python
harrold.find_food()
````

````python
harrold.dancing()
````

````python
harrold.eat_leaves()
````

<mark>Ще одна засада</mark>, треба переписуввати весь клас заново, якщо хочеш змінити щось незначне, інакше — помилка.

## Ініціалізація об'єкта

Іноді, створюючи об'єкт, ми хочемо задати деякі значення (_властивості, атрибути_) на потім.
Коли ми _ініціалізуємо_ об'єкт, ми готуємо його до використання.

Напр. ми хочемо задати кількість плям на жирафах коли створюємо (_ініуіалізуємо_). Для цього треба ств. ф-цію `__init__`. Це _особливий вид_ ф-ції, щоб задати властивості об'єкту під час його створення:

````python
class giraffe:
    def __init__(self, spots):
        self.giraffe_spot = spots
````

Так само, як і інші ф-ції, `__init__` мусить приймати першим параетром `self`. Далі ми присвоюємо змінній об'єкту значання параметра `spots`, застосовуючи параметр `self` --> `self.giraffe_spot = spots`.
Тобто: «Візьми значення параметру `self` та збережи його на потім (користуючись змінною об'єкту `giraffe_spot`)».
Так само як одна ф-ція з класу може викликати іншу ф-цію, застосовуючи параметр `self`, через нього ж можна отримати до змінних у класі.

````python
oswald = giraffe(100)
hertruda = giraffe(150)

print(oswald.giraffe_spots)
print(hertruda.giraffe_spots)
````

Цирк, якийсь. Мій код не працює, а Бріггса — так 👀

````python
class Giraffes:
    def __init__(self, spots):
        self.giraffe_spots = spots

ozwald = Giraffes(100)
gertrude = Giraffes(150)
print(ozwald.giraffe_spots)
print(gertrude.giraffe_spots)
````

Мабуть помилка була в якихось схожих іменах, типу: `giraffe` та `giraffe_spots`. Спробував використовувати різні імена і все працює:

```python
>>> class Turtles:
...     def __init__(self, eyes):
...         self.turtle_eye = eyes
...
>>> Thorin = Turtles(3)
>>> print(Thorin.turtle_eye)
3
```

Коли ми створюємо екзампляр `Thorin` класу `Turtles`, застосовуючи значення параметра `3`, ця дія викликає ф-цію `__init__` зі значенням `3` для параметру `eyes`.

Пам'ятай, створюючи об'єкт класу ми можемо звернутися до його змінних чи ф-цій, застосовуючи оператор-крапку та назву змінної або ф-ції, яку ми хочемо застосувати.
Але коли ми створюємо ф-ції всередині класу, ми звертаємося до тих же самих змінних/ф-цій, застовуючи параметр `self` --> `self.turtle_eye`.

### №1. Танець жирафи
Додай ф-ції до класу `Giraffes`, щоб пересунути ліву та праву ногу вперед та назад. Ф-ція для пересування лівої ноги вперед може виглядати ось так:

```python
def ліву_ногу_вперед(self):
    print('ліва нога вперед')
```

Потім створи ф-цію під назвою `танцювати`, щоб навчити Реджиналда танцювати.

````python
class Giraffes:
    def left_foot_forward(self):
        print('ліва нога вперед')
    def left_foot_backward(self):
        print('ліва нога назад')
    def right_foot_forward(self):
        print('права нога вперед')
    def right_foot_backward(self):
        print('права нога назад')
    def dancing(self):
        self.left_foot_forward()
        self.left_foot_backward()
        self.right_foot_forward()
        self.right_foot_backward()
        self.left_foot_backward()
        self.right_foot_backward()
        self.right_foot_forward()
        self.left_foot_forward()
        
````

````python
reginald = Giraffes()
reginald.dancing()
````
