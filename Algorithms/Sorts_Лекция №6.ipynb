## Квадратичные сортировки ($O(N^2)$)

Количество оперций на обработку массива равно $N^2$, где $N$ — длина массива.

**Симптотика алгоритма**:
- работа с памятью ($N^2$ памяти)
- скорости вычислений

### Вставками, Insertion sort, [Сортування включенням](https://uk.wikipedia.org/wiki/%D0%A1%D0%BE%D1%80%D1%82%D1%83%D0%B2%D0%B0%D0%BD%D0%BD%D1%8F_%D0%B2%D0%BA%D0%BB%D1%8E%D1%87%D0%B5%D0%BD%D0%BD%D1%8F%D0%BC).


Сравнения росли:
$$
1+2+3+4\approx \frac{N\times(N-1)}{2}\times k
$$

```python
A = [4, 2, 5, 1, 3]
```

```python
N = len(A)
for top in range(1, N):
    k = top
    while k > 0 and A[k - 1] > A[k]:
        A[k], A[k - 1] = A[k - 1], A[k]
        k -= 1
```

```python
print(A)
```

### Choise sort, Выбора

Та же самая _расширяющаяся прогрессия_

```python
A = list(range(10, 20)) + list(range(0, 10)); A
```

```python
N = len(A)
for pos in range(0, N - 1):
    for k in range(pos + 1, N):
        if A[k] < A[pos]:
            A[k], A[pos] = A[pos], A[k]
```

```python
print(A)
```

### Bubble sort, Пузырька

Сравнений на 1-м, 2-м ... проходе:
$$
4+3+2+1 = 10_{сравнений}
$$
Перед нами _арифметическая прогрессия:_
$$
1+2+..+(N-1_{проход})=\frac{N\times(N-1)}{2}=N^2
$$

```python
A = [4, 2, 4, 2, 1]
```

```python
N = len(A)
for bypass in range(1, N):
    for k in range(0, N - bypass):
        if A[k] > A[k + 1]:
            A[k], A[k + 1] = A[k + 1], A[k]
```

```python
print(A)
```
