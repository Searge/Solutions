````python
def max2(x, y):
    if x > y:
        return x
    elif x < y:
        return y
    else:
        print('They equal')
        
max2(5, 6)
````

````python
max2(5, 5)
````

### Решение учителя:
```python
def max2(x, y):
    if x > y:
        return x
    return y
```

_При равенстве просто возвращает цифру._

````python
def max3(x, y, z):
    return max2(x,  max2(y, z))
````

````python
max3(6, 4, 5)
````

````python
max3(8, 8, 7)
````

````python
max3(7, 8, 8)
````

> _От тут, якраз і вилізла помилка :(_

В Python есть правило **Duck typing** — _утиный полиморфизм_, если что-то выглядит, как утка, крякает как утка, значит это утка. Т.е. если что-то выглядит как строки, значит это строки.

Если тыпы в функции прямо неузказаны, значит, все, что можно сравнить — будет сравненно.

````python
def hello_separated(name='World', separator='-'):
    print('Hello,', name, sep=separator)
    
````

````python
hello_separated('John', '***')
````

````python
hello_separated(separator='***')
````

````python
hello_separated(separator='--',
               name='Jon')
````

## Структурное программирование

Згідно з цією методологією будь-яка програма може бути створена використовуючи три конструкції:
- **послідовне виконання** — одноразове виконання операції в порядку запису їх (операцій) в тексті програми;
- **розгалуження** — виконання певної операції або декількох операцій в залежності від стану певної, наперед заданої умови;
- **цикл** — багаторазове виконання операції або групи операцій за умови виконання деякої наперед заданої умови. Така умова називається — умова продовження циклу.

Кожна конструкція являє собою блок із одним входом і одним виходом.

Блок **Слідування** передбачає послідовне (лінійне) виконання операторів програми.

Блок **Вибір** являє собою точку прийняття рішення про подальший перебіг виконання операторів програми. Вибір здійснюється однією із трьох структур:
- `if` (єдиний вибір)
- `if…else` (подвійний вибір)
- `switch` або `case` (множинний вибір)

Усі три структури можна звести до одного типу `if`.

Блок **Повторення** реалізується одним із трьох способів:
- структура `while`
- структура `do/while`
- структура `for`

Усі три структури можна звести до структури `while`.

Структурована програма складається із вищеназваних блоків за двома правилами: **пакетування** (вихід одного блоку з'єднується із входом наступного) і **вкладення** (будь-який блок може бути замінений на структуру керування вибору або повторення)

https://uk.wikipedia.org/wiki/Структурне_програмування

## Метод грубой силы (brute force)
> Метод «грубої сили» (від англ. Brute force; або повний перебір) — метод рішення криптографічної задачі шляхом перебору всіх можливих варіантів ключа. Складність повного перебору залежить від кількості всіх можливих рішень задачі. Якщо простір рішень дуже великий, то повний перебір може не дати результатів протягом декількох років або навіть століть.
Будь-яка задача з класу NP може бути вирішена повним перебором. При цьому, навіть якщо обчислення цільової функції від кожного конкретного можливого рішення задачі може бути здійснена за поліноміальний час, в залежності від кількості всіх можливих рішень повний перебір може зажадати експоненціального часу роботи.
https://uk.wikipedia.org/wiki/Метод_«грубої_сили»

````python
def is_simple_number(x):
    """ Определяет, является ли число простым.
        x — целое положительное число.
        Если простое, то возвращает `True`,
        а иначе — `False`.
    """
    divisor = 2
    while divisor < x:
        if x%divisor == 0:
            return False
        divisor += 1
    return True
````

````python
is_simple_number(20)
````

````python
is_simple_number(4)
````

````python
is_simple_number(19)
````

````python
def factorize_number(x):
    """ Раскладывает число на множители.
        Печатает их на экран.
        x — целое положительное число.
    """
    divisor = 2
    while x > 1:
        if x%divisor == 0:
            print(divisor)
            x //= divisor
        else:
            divisor +=1
````
