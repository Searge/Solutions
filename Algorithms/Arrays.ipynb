# Массивы (тип: list)

`A -->` \[| 1 | 2 | 3 | 4 | 5 |]

```python
A = [1, 2, 3, 4, 5]

for x in A:
    print(x)
```

```python
for x in A:
    print(x, type(x))
    x += 1
    print(x)
```

```python
for k in range(5):
    A[k] = A[k] * A[k]
    
print(A)
```

```python
B = [0] * 1000
top = 0
x = int(input())
while x != 0:
    B[top] = x
    top += 1
    x = int(input())
```

```python
for k in range(13, -1, -1):
    print(B[k])
```

```python
N = int(input())

A = [0]*N
B = [0]*N

for k in range(N):
    A[k] = int(input())
    
```

```python
print(A, B)
```

```python
# copying from A to B

for k in range(N):
    B[k] = A[k]
    
print(A, B)
```
