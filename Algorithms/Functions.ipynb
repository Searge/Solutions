# Функции

````python
# простая, процедурная ф-ция

def hello():
    print('Hello, world')
    
hello()
````

````python
hello
````

````python
f = hello
# `f` становится ссылкой на ту же самую ф-цию
f()
````

````python
# формальный ↓параметр
def hello(name):
    print('Hello,', name)

hello('Searge')
````

`'Searge'` _→ фактический параметр._

````python
# задаем значение параметра по умолчанию
def hello(name = 'World'):
    print('Hello,', name + '!')

hello('Searge')
hello()
````

````python
def max2(x, y):
    if x > y:
        return x
    elif x < y:
        return y
    else:
        print('They equal')
        
max2(5, 6)
````

````python
max2(5, 5)
````

### Решение учителя:
```python
def max2(x, y):
    if x > y:
        return x
    return y
```

_При равенстве просто возвращает цифру._

````python
def max3(x, y, z):
    return max2(x,  max2(y, z))
````

````python
max3(6, 4, 5)
````

````python
max3(8, 8, 7)
````

````python
max3(7, 8, 8)
````

> _От тут, якраз і вилізла помилка :(_

В Python есть правило **Duck typing** — _утиный полиморфизм_, если что-то выглядит, как утка, крякает как утка, значит это утка. Т.е. если что-то выглядит как строки, значит это строки.

Если тыпы в функции прямо неузказаны, значит, все, что можно сравнить — будет сравненно.

````python
def hello_separated(name='World', separator='-'):
    print('Hello,', name, sep=separator)
    
````

````python
hello_separated('John', '***')
````

````python
hello_separated(separator='***')
````

````python
hello_separated(separator='--',
               name='Jon')
````
