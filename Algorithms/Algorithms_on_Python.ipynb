## Цикл for

```python
flag = False

# хотим узнать число, которое делится на 10
```

```python
N = int(input())

for i in range(N):# 1 число = стоп; т.е. от нуля дло N - 1
    x = int(input())
    if x % 10 == 0:
        flag = True
    print(flag)
    break
```

```python
N = int(input())

for i in range(N):
    x = int(input())
    flag = (x % 10 == 0) or flag
    print(flag)
    break
```

```python
# Хотим проверить, что все числа делятся на 10

flag = True
# т.е. считаем, что изначально все числа деляться на 10
```

```python
N = int(input())

for i in range(N):
    x = int(input())
    flag = flag and x % 10 == 0
    # записываем то ↑, что хотим проверять для каждого
    # и получается, что высказывания логическим умножением
    # все переумножаются
    # а изначально, переумножаем с единичкой (flag = True)
    # если было бы False, то 0 перемножил бы все на 0
    print(flag)
    break
```

## Вложенные и последовательные if-ы

```python
x = int(input())

if x % 2 == 0:
    print('да')
if x % 3 == 0:
    print('ага')
```

```python
x = int(input())

if x == 1:
    print('One')
if x == 2:
    print('Two')
```

```python
x = int(input())

if x == 1 or x == 2:
    print('да')
```

```python
x = int(input())

if x % 2 == 0:
    if x % 3 == 0:
        print('на 6')
        
# эквивалентно ↓

if x % 2 == 0 and x % 3 == 0:
    print('on 6')
```

```python
x = int(input())

if x % 2 == 0:
    if x % 3 == 0:
        print('на 6')
else:
    print('ещё')
        
# эквивалентно ↓

if x % 2 == 0 and x % 3 == 0:
    print('on 6')
else:
    print('else')
```

## Каскадные условные конструкции

Если число принадлежит отрицательным числам, то надо напечатать букву $A$, если число $\{0..5\}$ (не включая 5), то требуется напечатать букву $B$, если оно $\{5..10\}$ то требуется напечатать букву $C$.
Если $> 5  до + \infty$ то требуется напечатать букву $D$.

```python
x = int(input())

if x < 0:
    print('A')
elif x<5:
    print('B')
elif x<10:
    print('C')
elif x>10:
    print('D')
else:
    raise ValueError('A very specific bad thing happened.')
```
