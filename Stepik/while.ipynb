# Цикл while

````python
n = int(input())
c = 1
while c <= n:
    print('*' * c)
    c += 1
````

````python
n = 15
stars = '*'
while len(stars) <= n:
    print(stars)
    stars += '*'
````

````python
# Сколько всего знаков * будет выведено после исполнения фрагмента программы:
i = 0
while i < 5:
    print('*')
    if i % 2 == 0:
        print('**')
    if i > 2:
        print('***')
    i += 1
````

````python
# сумма чисел от 3 до 7
# 3+4+5+6+7=25
a,b = 3,7
s = 0
i = a
while i <= b:
    s += i
    i += 1
print(s)
````

````python
print(sum(range(3,8)))
````

````python
print( (a + b) * (b - a + 1) // 2 )
````

Напишите программу, которая считывает со стандартного ввода целые числа, по одному числу в строке, и после первого введенного нуля выводит сумму полученных на вход чисел.

Sample Input 1:

```5
-3
8
4
0
```
Sample Output 1:

<samp>
14
    </samp>


````python
sum = 0
while True:
    n = int(input())
    if n == 0: break
    sum += n
print(sum)
````

````python
sum = 0;
x = int(input())
sum += x;
while x != 0:
  x = int(input())
  sum += x;
print(sum)
````

````python
x = int(input())
res = 0
while x:
    res += x
    x = int(input())
print(res)
````
